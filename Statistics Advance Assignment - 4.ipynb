```python
'''Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.'''
'''
type 1 error-

when we reject the null hypothesis but in reality it is true

example - Imagine a medical test for a rare disease where the null hypothesis is that a patient has no disease. 
If the test incorrectly indicates a patient has the disease when they are actually disease-free, it is a Type 1 error

type 2 error-

when we retain the null hypothesis but in reality it is false

example- Suppose a new drug is tested to determine if it is effective in curing a disease.
The null hypothesis is that the drug has no effect, while the alternative hypothesis is that the drug is effective. 
If the study fails to show the drug's effectiveness when it actually does work, it is a Type 2 error
          '''
```
